```python
def convolution(k, data):
    n,m = data.shape
    img_new = []
    for i in range(n-3):
        line = []
        for j in range(m-3):
            a = data[i:i+3,j:j+3]
            line.append(np.sum(np.multiply(k, a)))
        img_new.append(line)
    return np.array(img_new)
```

The result of convolution is closely related to the convolution kernel, which is a matrix.  
Though there are no obvious distinction between the original pictures, it is because of the convolution kernel.

```python
img_new = convolution(k, data)
plt.imshow(img_new, cmap='gray')
plt.axis('off')
```

### Try Different Convolution Kernels

```python
#Convolution Kernel for Vertical Edge Detection
k1 = np.array([
    [1,0,-1],
    [1,0,-1],
    [1,0,-1]
])
img_new = convolution(k1, data)
plt.imshow(img_new, cmap='gray')
plt.axis('off')
```

We can see that the object within the image's vertical features are extracted, and other features are blurred.  
This is because the convolution kernel is a vertical matrix.  
And this is very important while processing image data for your training model.  

```python
k2 = np.array([
    [1,1,1],
    [0,0,0],
    [-1,-1,-1]
]) #Convolution Kernel for Horizontal Edge Detection
img_new = convolution(k2, data)
plt.imshow(img_new, cmap='gray')
plt.axis('off')
```
